# Working Class Of India

The working class of India is a large and diverse group of people. They are employed in a variety of sectors, including agriculture, manufacturing, construction, and services. The working class is also divided by gender, caste, and religion. The working class in India faces a number of challenges. They are often paid low wages, have poor working conditions, and lack access to social security benefits. The working class is also vulnerable to exploitation by employers and to discrimination by the government. Despite the challenges they face, the working class in India has a long history of struggle and activism. They have fought for better wages, working conditions, and social security benefits. They have also fought against discrimination and for social justice.

The working class in India is a powerful force for change. They have the potential to transform India into a more just and equitable society.

# 1 | Differnet Working Class 

The working class is generally devided into 

|______|______
|---|---
|Total Working Class|The total Working Class, comprises of both the currently working people and non-working people
|Main Working Class|The working class that is majorly working
|_3 Working Class|The working class that worked for less than 3 months
|6 Working Class|The working class that worked for around 3-6 months
|Seeking Working Class|The working class that is currently working but also seeking to do another work
|Non-Working Class|The working class that is currently not working
|Non Seeking Wokring Class|The working class that is not working but seeking to do work

Their ratios are 

```python
x = ["Main Workers Total Population" , "_3 Total Population" , "6 Total Population" , "Seeking Total Population" , "Non Seeking Total Population"]
labels = ["Main Workers" , "<3 Months" , "3-6 Months" , "Seeking" , "Non Working Seeking"]

values = [np.sum(general[c]) for c in x ]

explode = [np.sum(general[c])/100000000000 for c in x ]
colors = ["#37E2D5" , "#590696" , "#C70A80" , "#FBCB0A" , "#E21818"]

plt.pie(values, labels = labels , shadow = True , explode = explode , colors = colors , autopct='%1.1f%%')
plt.show() 
```

# 2 | Working Population by States 

```python
labels = ['West Bengal', 'Karnataka','Madhya Pradesh', 'Bihar','Rajasthan', 'Uttar Pradesh', 'Gujarat', 'Maharashtra', 'Andhra Pradesh','Tamil Nadu']

explode = [np.sum(general[general["States"] == state]["Total Population"])/10000000000 for state in labels]
values = [np.sum(general[general["States"] == state]["Total Population"]) for state in labels]

plt.pie(values , labels = labels , explode = explode , shadow = True , autopct='%1.1f%%')
plt.show()
```
